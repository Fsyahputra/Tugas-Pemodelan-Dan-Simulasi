```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x =  np.linspace(0, 100, 100)
y = 10*np.random.rand(len(x))

fig, ax = plt.subplots()

ax.scatter(x, y, label='Random Data', marker=',')
ax.set_title('Random Data Plot')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.grid()

plt.show()

```

```python
x = np.linspace(0, 10*np.pi, 200)
y = np.cos(x)
fig, ax = plt.subplots()
ax.plot(x,y)
ax.set_title('Grafik Cosinus')
ax.set_xlabel('x')
ax.set_ylabel('y = cos(x)')
plt.grid()
plt.show()
```
